## Read and Store the Tally Results

```python
import os
import pandas as pd
from GeneralNuclear.MCNP import read_tally

# Switch to foil directory
os.chdir('Foil')  
```

```python
# Energy index map (from worm input)
energy = [843.76, 1014.52, 1368.63, 355.7, 411.8, 336.241, 1293.56, 1377.63, 810.76, 909.15]
x = [0] 
y = [0] 

# Create dataframe
print "Currently reading from: {}".format(os.getcwd())
foil = pd.DataFrame(columns=['energy', 'x', 'y', 'z', 'modelEff', 'modelUncert'], dtype=float)

for filename in os.listdir(os.getcwd()):
    splitFile = filename.split('.')
    if splitFile[-1] == 'o':
        print filename
        splitFile = filename.split('_')
        tmpTally = read_tally(os.getcwd()+'/'+filename, 8, readGroups=True)[0]
        
        # Loop over tally to find the data corresponding to the peak and store
        for index, row in tmpTally.iterrows():
            if float(row[0]) > energy[int(splitFile[2])-1]/1000.:
                tally = float(row[1])
                uncert = float(row[2])
                break
                
        foil = foil.append(pd.Series([energy[int(splitFile[2])-1], x[int(splitFile[3])-1], y[int(splitFile[3])-1], 
                                        float(splitFile[4].split('.')[0]), tally, tally*uncert], 
                                       index=['energy', 'x', 'y', 'z', 'modelEff', 'modelUncert']), ignore_index=True)

foil
```

```python
# Create dataframe
point = pd.DataFrame(columns=['energy', 'x', 'y', 'z', 'modelEff', 'modelUncert'], dtype=float)

print "Currently reading from: {}".format(os.getcwd())
for filename in os.listdir(os.getcwd()):
    splitFile = filename.split('.')
    if splitFile[-1] == 'o':
        print filename
        splitFile = filename.split('_')
        tmpTally = read_tally(os.getcwd()+'/'+filename, 8, readGroups=True)[0]
        
        # Loop over tally to find the data corresponding to the peak and store
        for index, row in tmpTally.iterrows():
            if float(row[0]) > energy[int(splitFile[2])-1]/1000.:
                tally = float(row[1])
                uncert = float(row[2])
                break
                
        point = point.append(pd.Series([energy[int(splitFile[2])-1], x[int(splitFile[3])-1], y[int(splitFile[3])-1], 
                                        float(splitFile[4].split('.')[0]), tally, tally*uncert], 
                                       index=['energy', 'x', 'y', 'z', 'modelEff', 'modelUncert']), ignore_index=True)

point
```

## Geometry Correction Factor @1 cm

```python
from GeneralNuclear.BasicNuclearCalcs import fractional_solid_angle
from GeneralNuclear.Counting import volume_solid_angle
from math import sqrt

print "Distance Efficiency Scaling Factor:"
print "@1cm: ", fractional_solid_angle(3.245, 1.595)
print "@18cm: ", fractional_solid_angle(3.245, 18.595)
print "S_18:", fractional_solid_angle(3.245, 18.595)/fractional_solid_angle(3.245, 1.595)

print "\nAnalytic Geometry Correction Factor between pt src at 1 cm and Volume Source:"
print "Eff @1cm pt: {:.3f}".format(fractional_solid_angle(3.245, 1.595))
print "Eff @1cm vol: {:.3f}".format(volume_solid_angle(2.5, 3.245, 1.595))
print "F_1: {:.3f}".format(volume_solid_angle(2.5, 3.245, 1.595)/fractional_solid_angle(3.245, 1.595))

print "\nModeled Geometry Correction Factor between pt src at 1 cm and Volume Source for Al 1014.52 keV line:"
print "Eff @1cm pt: {:.3f}".format(point.loc[4]['modelEff'])
print "Eff @1cm vol: {:.3f}".format(foil.loc[4]['modelEff'])
print "F_1: {:.3f} +/- {:.3f}".format(foil.loc[4]['modelEff']/point.loc[4]['modelEff'],
                               sqrt((foil.loc[4]['modelUncert']/foil.loc[4]['modelEff'])**2
                                    +(point.loc[4]['modelUncert']/point.loc[4]['modelEff'])**2)*
                               foil.loc[4]['modelEff']/point.loc[4]['modelEff'])

print "\nModeled Geometry Correction Factor between pt src at 1 cm and Volume Source for In 336.52 keV line:"
print "Eff @1cm pt: {:.3f}".format(point.loc[12]['modelEff'])
print "Eff @1cm vol: {:.3f}".format(foil.loc[12]['modelEff'])
print "F_1: {:.3f} +/- {:.3f}".format(foil.loc[12]['modelEff']/point.loc[12]['modelEff'],
                               sqrt((foil.loc[12]['modelUncert']/foil.loc[12]['modelEff'])**2
                                    +(point.loc[12]['modelUncert']/point.loc[12]['modelEff'])**2)*
                               foil.loc[12]['modelEff']/point.loc[12]['modelEff'])
```

```python
# Create dataframe
fGamma = pd.DataFrame(columns=['energy', 'f'], dtype=float)

for i in range(0, len(foil)):
    if foil.iloc[i]['z'] != 18:
        fGamma = fGamma.append(pd.Series([foil.iloc[i]['energy'], foil.loc[i]['modelEff']/point.loc[i]['modelEff']], 
                               index=['energy', 'f']), ignore_index=True)
fGamma.sort_values('energy')
```
